```python
sns.displot(data['sex'])
plt.xticks([0,1],['female','male'])
plt.show()
```

**30% of the patients are female**

```python
sns.countplot(x='sex',hue='target',data=data)
plt.xticks([0,1],['female','male'])
plt.legend(labels=['disease','no disease'])
plt.show()
```

## **the ratio of diseased to non diseased is more in females**

```python
sns.countplot(x='cp',data=data)
```

**chest pain type 0 is more common in the people**

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['disease','no disease'])
plt.show()
```

**people having chest pain type '0' are rarely diseased whereas people with chest pain type '2' have more chances of being categorized as diseased.**

```python
g=sns.FacetGrid(data,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',fill=True)
plt.legend(labels=['male','female'])
plt.show()
```

**females have 120 as the highest blood pressure as compared tom males having bp of 140 as maximum**
